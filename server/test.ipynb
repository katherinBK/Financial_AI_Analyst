```python
from finvizfinance.quote import finvizfinance

def get_forex_data(pair):
    """
    Fetch Forex data for a specific currency pair from Finviz.
    
    Args:
        pair (str): The currency pair (e.g., 'EURUSD', 'GBPUSD').
    
    Returns:
        dict: A dictionary containing Forex data.
    """
    try:
        # Initialize the Finviz object with the currency pair
        forex = finvizfinance(pair)
        
        # Get the Forex data
        data = forex.ticker_fundament()
        
        # Print the data
        print(f"Forex Data for {pair}:")
        for key, value in data.items():
            print(f"{key}: {value}")
        
        return data
    except Exception as e:
        print(f"Error fetching Forex data: {e}")
        return None

# Example usage
if __name__ == "__main__":
    pair = "EURUSD"  # Example currency pair
    forex_data = get_forex_data(pair)
```
